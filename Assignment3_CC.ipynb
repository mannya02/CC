## Q6) Create a sample dataset (employees.csv) containing information about employees in a company.

```python
data1 = {
    "Employee_ID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Edward"],
    "Department": ["HR", "IT", "IT", "Marketing", "Sales"],
    "Age": [29, 34, 41, 28, 38],
    "Salary": [50000, 70000, 65000, 55000, 60000],
    "Years_of_Experience": [4, 8, 10, 3, 12],
    "Joining_Date": ["2020-03-15", "2017-07-19", "2013-06-01", "2021-02-10", "2010-11-25"],
    "Gender": ["Female", "Male", "Male", "Female", "Male"],
    "Bonus": [5000, 7000, 6000, 4500, 5000],
    "Rating": [4.5, 4.0, 3.8, 4.7, 3.5]
}

data2 = pd.DataFrame(data1)
data2.to_csv("employees.csv", index=False)
print("Dataset created and saved as 'employees.csv'")
```

## WAP to determine the following using dataset in Q6:
a) Shape (number of rows and columns) of the DataFrame.

b) Summary of the DataFrame that includes the data types and non-null counts for
each column.

c) Generate descriptive statistics.

d) Display the first 5 rows and last 3 rows of the dataset.

e) Calculate the following statistics from the dataset:
i. The average salary of employees.
ii. The total bonus paid to all employees.
iii. The youngest employee's age.
iv. The highest performance rating.

f) Sort the DataFrame by the Salary column in descending order.

g) Add a new column that categorizes employees based on their performance rating:

i. Excellent for ratings >= 4.5
ii. Good for ratings >= 4.0 but < 4.5
iii. Average for ratings < 4.0
h) Identify missing values in the DataFrame.
i) Rename the Employee_ID column to ID.
j) Find all employees who:

i. Have more than 5 years of experience.
ii. Belong to the IT department.

k) Modify the dataset by adding a new column, Tax, which deducts 10% of the
Salary.

l) Save the modified DataFrame (with added columns) to a new CSV file.

```python
# (a) Shape (number of rows and columns) of the DataFrame.
s=data2.shape
print('(a) Shape=',s)

# (b) Summary of the DataFrame
print("\n(b) Summary of the DataFrame:")
print(data2.info())

# (c) Descriptive statistics
print("\n(c) Descriptive statistics:")
print(data2.describe())

# (d) Display first 5 and last 3 rows
print("\n(d) First 5 rows:")
print(data2.head())

print("\nLast 3 rows:")
print(data2.tail(3))

# (e) Calculate the following statistics from the dataset:
# i. The average salary of employees.
# ii. The total bonus paid to all employees.
# iii. The youngest employee's age.
# iv. The highest performance rating.
avg_salary = data2["Salary"].mean()
total = data2["Bonus"].sum()
youngest_age = data2["Age"].min()
highest_rating = data2["Rating"].max()

print("\n(e) Statistics:")
print("i. Average Salary:", avg_salary)
print("ii. Total Bonus Paid:", total)
print("iii. Youngest Employee Age:", youngest_age)
print("iv. Highest Performance Rating:", highest_rating)

# (f) Sort by Salary in descending order
data2_sorted = data2.sort_values(by="Salary", ascending=False)
print("\n(f) Sorted DataFrame by Salary (Descending):")
print(data2_sorted)

# (g) Add a new column that categorizes employees based on their performance rating:
# i. Excellent for ratings >= 4.5
# ii. Good for ratings >= 4.0 but < 4.5
# iii. Average for ratings < 4.0
def categ_rating(rating):
    if rating >= 4.5:
        return "Excellent"
    elif rating >= 4.0:
        return "Good"
    else:
        return "Average"

data2["Performance_Category"] = data2["Rating"].apply(categ_rating)
print("\n(g) DataFrame with Performance Category:")
print(data2)

# (h) Identify missing values in the DataFrame.
print("\n(h) Missing values in the DataFrame:")
print(data2.isnull().sum())

# (i) Rename 'Employee_ID' column to 'ID'
data2.rename(columns={"Employee_ID": "ID"}, inplace=True)
print("\n(i) Renamed DataFrame:")
print(data2.head())

# (j) Find employees with:
# i. More than 5 years of experience
exp_above_5 = data2[data2["Years_of_Experience"] > 5]
print("\n(j-i) Employees with more than 5 years of experience:")
print(exp_above_5)

# ii. Belonging to the IT department
it_employees = data2[data2["Department"] == "IT"]
print("\n(j-ii) Employees in IT Department:")
print(it_employees)

# (k) Add Tax Column (10% deduction)
data2["Tax"] = data2["Salary"] * 0.10
print("\n(k) DataFrame with Tax column:")
print(data2)

# (l) Save modified DataFrame to a new CSV file
data2.to_csv("employees_modified.csv", index=False)
print("\n(l) Modified DataFrame saved as 'employees_modified.csv'")
```
