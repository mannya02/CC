#### Q1.) For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), Find (i) Sum of all elements (ii) Sum of all elements row-wise (iii) Sum of all elements column-wise

```python
import numpy as np
gfg=np.matrix('[4,1,9;12,3,1;4,5,6]')
total = np.sum(gfg)
print("Total Sum=",total)

row_sum=np.sum(gfg,axis=1)
print("Row wise Sum=",row_sum)

col_sum=np.sum(gfg,axis=0)
print("Column wise Sum=",col_sum)

```

#### Q2.) (a)For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find (i) Sorted array (ii) Indices of sorted array (iii) 4 smallest elements (iv) 5 largest elements (b) For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find (i) Integer elements only (ii) Float elements only

```python
array = np.array([10, 52, 62, 16, 16, 54, 453])
sort_array = np.sort(array)
print("Sorted array:", sort_array)

sort_indices = np.argsort(array)
print("Indices of sorted array:", sort_indices)

smallest_4 = np.sort(array)[:4]
print("4 smallest elements:", smallest_4)

largest_5 = np.sort(array)[-5:]
print("5 largest elements:", largest_5)

array2 = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

int_elements = array2[array2 == np.floor(array2)]
print("Integer elements:", int_elements)

float_elements = array2[array2 != np.floor(array2)]
print("Float elements:", float_elements)
```

#### Q3.) Q.3 You are given a weekly sales dataset and need to perform various operations using NumPy broadcasting. a) Generate your unique sales dataset: Take the sum of the ASCII values of the initials of your first and last name. Call this value X. (If your initials are A B → ASCII sum = 65 + 66 = 131 → sales = [131, 181, 231, 281, 331].) Create a NumPy array sales with values [X, X+50, X+100, X+150, X+200]. b) Compute your personalized tax rate as ((X % 5) + 5) / 100. Use broadcasting to apply this tax rate to each sales value. c) Adjust sales based on discount: If sales < X+100, apply a 5% discount. If sales >= X+100, apply a 10% discount. d) Expand sales data for multiple weeks:  Create a 3×5 matrix representing three weeks of sales by stacking sales three times using broadcasting. Increase sales by 2% per week using element-wise broadcasting.

```python
first_initial = 'M'  
last_initial = 'S'   

X = ord(first_initial) + ord(last_initial)
sales = np.array([X, X+50, X+100, X+150, X+200])
print("Sales dataset:", sales)
tax_rate = ((X % 5) + 5) / 100
after_tax = sales * (1 - tax_rate)
print("Sales after tax:", after_tax)
dis_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)
print("Discounted sales:", dis_sales)
weeks_sales = np.vstack([sales, sales * 1.02, sales * 1.04])
print("Sales for 3 weeks with 2% increment per week:\n", weeks_sales)
```

#### Q4.)Generate x values using np.linspace() from -10 to 10 with 100 points. Use each function from the list below and compute y values using NumPy: (i)Y = x^2 (ii) Y = sin(x)  (iii)Y = e^x  (iv)Y = log(|x| + 1) .Plot the chosen function using Matplotlib. Add title, labels, and grid for clarity.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)

y1 = x ** 2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)

plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.plot(x, y1, label="y = x^2", color='b')
plt.title("y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(x, y2, label="y = sin(x)", color='r')
plt.title("y = sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

plt.subplot(2, 2, 3)
plt.plot(x, y3, label="y = e^x", color='g')
plt.title("y = e^x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

plt.subplot(2, 2, 4)
plt.plot(x, y4, label="y = log(|x| + 1)", color='m')
plt.title("y = log(|x| + 1)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

plt.tight_layout()
plt.show()
```
